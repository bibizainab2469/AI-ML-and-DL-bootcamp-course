#### while Loop

```python
count = 1
while count <= 10:
    print(count,end= " ")
    count += 1
```

```python
count = 10
while count >= 1:
    print(count, end = " ")
    count -= 1
```

```python
i = 1
total = 0
while i <= 10:
    total += i
    i += 1
print("Sum:", total)
```

```python
i = 1
n = 6
while i < 11:
    print(n,"*",i,"=",i*n)
    i += 1
```

```python
number = int(input("Enter a number"))
i = 1
fact = 1
while number >= i:
    fact *= i
    i += 1
print("factorial of ",number,":",fact)
```

```python
digit = int(input("Enter a number")) # sum of digits in a number
total = 0
while digit > 0:
    total += digit%10
    digit//= 10
print("digit:",total)
```

#### Nested While

```python
i = 1
while i <= 3:
    j = 1
    while j <= 3:
        print(i, j)
        j += 1
    i += 1
```

```python
i = 1
while i <= 3:
    j = 1
    while j <= 4:
        print("(",i,j,")", end=" ")
        j += 1
    print()
    i += 1
```

```python
count = 1
while count <= 3:
    for i in range(1, 6):
        print(i, end=" ")
    print()
    count += 1
```

```python
numbers = [2, 3, 4]
repeat = 1
while repeat <= 2:
    for num in numbers:
        print(num * repeat, end=" ")
    print()
    repeat += 1
```
