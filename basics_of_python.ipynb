# Casting

From other dtypes to int

```python
x = int(4)
y = int(4.76)
z = int("6")
a = int(True)
print(type(x),type(y), type(z),type(a), sep="\n")
print(a)
```

from dtyps to float

```python
x = float(4)
y = float(4.76)
z = float("6")
a = float(True)
print(type(x),type(y), type(z),type(a), sep="\n")
print(a)
```
